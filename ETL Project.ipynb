```python
def match_term(term, list_names, min_score=0):

    max_score = -1
    
    max_name = "no match"
    
    for term2 in list_names:
        score = fuzz.ratio(term, term2)
        
        if (score > min_score) & (score > max_score):
            max_name = term2
            max_score = score
    return(max_name, max_score)
        
```

```python
for i in products_prices_Chedraui:
    print(i,match_term(i,products_prices_Soriana,80))
```

```python
dict_list = []

for name in products_prices_Chedraui:
    
    match = match_term(name,products_prices_Soriana,80)
    
    dict_ = {}
    dict_.update({"Chedraui": name})
    dict_.update({"Soriana": match[0]})
    dict_.update({"score": match[1]})
    dict_list.append(dict_)
```

```python
dict_list
```

```python
import numpy as np 
```

```python
index=np.arange(len(dict_list))
```

```python
comparative_table=pd.DataFrame({"index":index})
comparative_table["chedraui_product"]=""
comparative_table["chedraui_price"]=""
comparative_table["soriana_product"]=""
comparative_table["soriana_price"]=""
comparative_table["score"]=""
del comparative_table['index']
comparative_table.head(2)
```

```python

for index, row in comparative_table.iterrows():   
    try:
        comparative_table.at[index,'chedraui_product'] = dict_list[index]['Chedraui'][0]
        comparative_table.at[index,'chedraui_price'] = dict_list[index]['Chedraui'][1].split('$')[1]
        comparative_table.at[index,'soriana_product']=dict_list[index]['Soriana'][0]
        comparative_table.at[index,'soriana_price']=dict_list[index]['Soriana'][1].split('$')[1]
        comparative_table.at[index,"score"]=dict_list[index]['score']
    except (KeyError, IndexError, ValueError):
        print(f"{index} not found. Skipping...")
```

```python
comparative_table.head(20)
```

```python
comparative_table["chedraui_product"]=comparative_table["chedraui_product"].astype('str')
comparative_table["chedraui_price"]=pd.to_numeric(comparative_table.chedraui_price)
comparative_table["soriana_product"]=comparative_table["soriana_product"].astype('str')
comparative_table["soriana_price"]=pd.to_numeric(comparative_table.soriana_price)
comparative_table["score"]=pd.to_numeric(comparative_table.score)
print(comparative_table.dtypes)
```

```python
from database import host,port,username,password,db
from sqlalchemy import create_engine, Column, Integer, String, func, Date, DateTime,Float,ForeignKey
from sqlalchemy.orm import relationship
import psycopg2
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
```

```python
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}/{db}")
Base_1 = declarative_base(bind=engine)
```

```python
class Comparative(Base_1):
    __tablename__ = 'comparative_table'
    id= Column(Integer,primary_key=True, nullable=False)
    chedraui_product = Column(String,nullable=False)
    chedraui_price = Column(Float)
    soriana_product = Column(String)
    soriana_price = Column(Float)
    score= Column(Float)
```

```python
Base_1.metadata.tables
```

```python
Base_1.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(engine)
```

```python
for x in range(len(comparative_table)):
    session.add(Comparative(chedraui_product=comparative_table["chedraui_product"][x], 
                            chedraui_price=comparative_table["chedraui_price"][x],
                            soriana_product=comparative_table["soriana_product"][x],
                            soriana_price=comparative_table["soriana_price"][x],
                            score=comparative_table["score"][x]))
```
