```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import pandas as pd
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

## Chedraui

```python
url_Chedraui = "https://www.chedraui.com.mx/Departamentos/S%C3%BAper/Frutas-y-Verduras/Frutas/c/MC210101?siteName=Sitio+de+Chedraui&isAlcoholRestricted=false"

products_prices_Chedraui =[]

browser.visit(url_Chedraui)

browser.html

soup = bs(browser.html, 'html.parser')

divs = soup.find_all('ul', class_='pagination')

numbers= list(range(len(divs[0].find_all('li'))-3))                                  
                                        
for x in numbers:
                                        
    url =(f"https://www.chedraui.com.mx/Departamentos/S%C3%BAper/Frutas-y-Verduras/Frutas/c/MC210101?q=%3Arelevance&page={x}&pageSize=200")

    browser.visit(url)

    browser.html

    soup = bs(browser.html, 'html.parser')

    containers = soup.find_all("li", class_= "js-plp-product-click product-item")
#     if len(containers) == 0:
#                 break
            
    for container in containers:

            product = container.a["title"].strip()

            price = container.findAll("div", {"class": "product__listing--price price-colour-final"})[0].text.strip()
            
            products_prices_Chedraui.append((product,price))
            
```

## Soriana

```python
url_Soriana = "https://superentucasa.soriana.com/default.aspx?p=13287&px=1&nuor=0"

browser.visit(url_Soriana)

browser.html

soup = bs(browser.html, 'html.parser')

divs = soup.find_all('ul', class_='pagination')

products_prices_Soriana =[]

numbers= list(range(len(divs[0].find_all('li'))))                                  
numbers = numbers[1:]                                       
for x in numbers:
                                        
    url_Soriana=(f"https://superentucasa.soriana.com/default.aspx?p=13287&px={x}&nuor=0")

    browser.visit(url_Soriana)
    
    browser.html

    soup = bs(browser.html, 'html.parser')

    containers = soup.find_all("div", class_= "col-lg-3 col-md-4 col-sm-12 col-xs-12 product-item")
    containers =  containers [1:]
    
    if len(containers) == 0:
                break
            
    for container in containers:

            product = container.h4.text

            price = container.find('h4', class_ ="precio-regular-plp").text
            
            products_prices_Soriana.append((product,price))
            
```

```python
print(len(products_prices_Chedraui))
print(len(products_prices_Soriana))
```

```python
def match_term(term, list_names, min_score=0):

    max_score = -1
    
    max_name = "no match"
    
    for term2 in list_names:
        score = fuzz.ratio(term, term2)
        
        if (score > min_score) & (score > max_score):
            max_name = term2
            max_score = score
    return(max_name, max_score)
        
```

```python
for i in products_prices_Chedraui:
    print(i,match_term(i,products_prices_Soriana,80))
```

```python
dict_list = []

for name in products_prices_Chedraui:
    
    match = match_term(name,products_prices_Soriana,80)
    
    dict_ = {}
    dict_.update({"Chedraui": name})
    dict_.update({"Soriana": match[0]})
    dict_.update({"score": match[1]})
    dict_list.append(dict_)
```
